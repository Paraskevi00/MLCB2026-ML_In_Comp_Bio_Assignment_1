```python
from scipy.stats import spearmanr
```

```python
# Set number of bootstrap iterations
n_iterations = 50
# Define subsample size (80% of training data)
subsample_size = int(0.8 * len(cpg_train))
# Number of top features to select
top_k = 200

# Convert arrays to DataFrames with feature names
cpg_train_df = pd.DataFrame(cpg_train, columns=numer_features)
cpg_val_df=pd.DataFrame(cpg_val,columns=numer_features)

# Initialize counter for feature selection frequency
feature_counts = {col: 0 for col in cpg_train_df.columns}
print(feature_counts)
```

```python
#Stability Selection
np.random.seed(42)

for i in range(n_iterations):
    
    indices = np.random.choice(len(cpg_train), subsample_size, replace=False)
    
    X_sub = cpg_train_df.iloc[indices]
    y_sub = y_train.iloc[indices]
    
    # Compute Spearman correlation for each feature
    correlations = []
    
    for col in X_sub.columns:
        corr, _ = spearmanr(X_sub[col], y_sub)
        corr = 0 if np.isnan(corr) else abs(corr)
        correlations.append(corr)
    
    correlations = np.array(correlations)
    
    # Get top 200 features
    top_indices = np.argsort(correlations)[-top_k:]
    top_features = X_sub.columns[top_indices]
    
    # Count selections
    for feat in top_features:
        feature_counts[feat] += 1
```

```python
# Assigning threshold for stable features (selected frequently)
threshold = n_iterations / 2  # > 25

stable_features = [feat for feat, count in feature_counts.items() if count > threshold]

print("Number of stable features:", len(stable_features))
print(stable_features)
```

```python
# Convert feature counts to DataFrame
freq_df = pd.DataFrame.from_dict(feature_counts, orient='index', columns=['count'])

# Compute selection frequency for each feature
freq_df['frequency'] = freq_df['count'] / n_iterations
# Sort features by frequency
freq_df = freq_df.sort_values(by="frequency", ascending=False)
print(freq_df.head(10))

print(freq_df.describe())
```

```python
#Histogram of feature selection frequencies
plt.figure(figsize=(8,5))
plt.hist(freq_df['frequency'], bins=20)
plt.xlabel("Selection Frequency")
plt.ylabel("Number of Features")
plt.title("Stability Selection Frequency Distribution")

plt.savefig("../figures/stability_selection_distribution.png")
plt.show()
```

```python
from feature_engine.selection import MRMR
from sklearn.metrics import mean_squared_error
```

```python
# Define different numbers of features to select
K_values = [50, 100, 200]
# Store mRMR results for each K
mrmr_results = {}
# Store selected feature subsets
selected_dict = {}
```

```python
for K in K_values:
    # method="FCQ": uses correlation-based feature selection
    sel = MRMR(method="FCQ", regression=True, max_features=K)

    # Fit the selector on training data
    sel.fit(cpg_train, y_train)
    
    # Transform training and validation data to keep only selected features
    X_train_sel = sel.transform(cpg_train)
    X_val_sel = sel.transform(cpg_val)
    
    model = LinearRegression()
    model.fit(X_train_sel, y_train)
    
    y_val_pred = model.predict(X_val_sel)
    
    # Compute RMSE
    rmse = np.sqrt(mean_squared_error(y_val, y_val_pred))
    
     # Store RMSE results for this K
    mrmr_results[K] = rmse
    # Store the fitted selector
    selected_dict[K] = sel
print(mrmr_results)
print(selected_dict)

#for k, v in mrmr_results.items():
    #print(f"K={k}, RMSE={v:.4f}")
```

```python
#Selecting best number of features
best_K = min(mrmr_results, key=mrmr_results.get)
best_selector = selected_dict[best_K]

print("Best K:", best_K)
```

```python
# Store the names of the selected features
selected_features = cpg_train_df.columns[best_selector.get_support()]
# Take the top 10 selected features
top10 = selected_features[:10]

# Compute Spearman correlation and create Dataframe to display their scores
scores = []
for feat in top10:
    corr = abs(spearmanr(cpg_train_df[feat], y_train)[0])
    scores.append(corr)

top10_df = pd.DataFrame({
    "Feature": top10,
    "Score (|Spearman|)": scores
})

print(top10_df)

selected_indices = np.where(best_selector.get_support())[0]
print(selected_indices)
```

```python
#Calcuate overlap between the two feature selection methods
stable_set = set(stable_features)
mrmr_set = set(selected_features)

overlap = stable_set.intersection(mrmr_set)

print("Overlap:", len(overlap))
```

```python
#Importing necessary libraries 
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import loguniform, uniform
```

```python
el_net=ElasticNet(max_iter=10000,tol=0.001)
# Define parameter distributions for Randomized Search
# alpha: regularization strength (log-uniform distribution over a wide range)
# l1_ratio: balance between L1 (Lasso) and L2 (Ridge), between 0.1 and 1.0
params_grid_net= {
    "alpha" : loguniform(0.001,10),
    "l1_ratio" : uniform(0.1,0.9)                        
}

# Set up RandomizedSearchCV for hyperparameter tuning
random_net=RandomizedSearchCV(
    el_net,
    params_grid_net,
    n_iter=40,
    cv=5,
    scoring="neg_root_mean_squared_error",
    random_state=42
)

# Fit the randomized search on training data
random_net.fit(cpg_train,y_train)

print("Best Parameters:", random_net.best_params_)
print("Best cross-validated rmse:", abs(random_net.best_score_))
rmse_random_en = abs(random_net.best_score_)
```

```python
svr=SVR()
# Define parameter distributions for Randomized Search
# C: regularization parameter (controls trade-off between smoothness and fitting)
# epsilon: margin of tolerance where no penalty is given to errors
# kernel: type of function used to map data (linear or nonlinear "rbf")
params_grid_svr={
    "C" : loguniform(0.1,500),
    "epsilon" : [0.01, 0.1, 0.5, 1.0],
    "kernel" : ["rbf","linear"]
}
# Set up RandomizedSearchCV for hyperparameter tuning
random_svr=RandomizedSearchCV(
    svr,
    params_grid_svr,
    n_iter=40,
    cv=5,
    scoring="neg_root_mean_squared_error",
    random_state=42
)
#Fit on SVR model
random_svr.fit(cpg_train,y_train)

print("Best Parameters:", random_svr.best_params_)
print("Best cross-validated rmse:", abs(random_svr.best_score_))
rmse_random_svr = abs(random_svr.best_score_)
```

```python
bayes=BayesianRidge()
# Define parameter distributions for Randomized Search
# alpha_1, alpha_2: hyperparameters controlling the prior on noise
# lambda_1, lambda_2: hyperparameters controlling the prior on weights
# Small values allow more flexibility; larger values increase regularization
params_grid_bayes={
    "alpha_1" : loguniform(1e-7,1e-3),
    "alpha_2" : loguniform(1e-7,1e-3),
    "lambda_1" : loguniform(1e-7,1e-3),
    "lambda_2" : loguniform(1e-7,1e-3)
}
# Set up RandomizedSearchCV for hyperparameter tuning
random_bayes=RandomizedSearchCV(
    bayes,
    params_grid_bayes,
    n_iter=40,
    cv=5,
    scoring="neg_root_mean_squared_error",
    random_state=42
)
#Fit on Bayesian Ridge model
random_bayes.fit(cpg_train,y_train)

print("Best Parameters:", random_bayes.best_params_)
print("Best cross-validated rmse:", abs(random_bayes.best_score_))
rmse_random_bayes = abs(random_bayes.best_score_)
```
