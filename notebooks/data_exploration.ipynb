```python
# Target
y_train = train_df["age"]
y_val = val_df["age"]

# Feature groups
cpg_cols = [c for c in train_df.columns if c.startswith("cg")]
meta_cols = ["sex", "ethnicity"]

# CpG only
X_train_cpg = train_df[cpg_cols]
X_val_cpg = val_df[cpg_cols]

# Metadata only
X_train_meta = train_df[meta_cols]
X_val_meta = val_df[meta_cols]

# Combined
X_train_all = train_df[cpg_cols + meta_cols]
X_val_all = val_df[cpg_cols + meta_cols]
```

```python
# CpG only
pre_cpg = build_preprocessor(cpg_cols, [])
X_train_cpg_p = pre_cpg.fit_transform(X_train_cpg)
X_val_cpg_p = pre_cpg.transform(X_val_cpg)

# Metadata only
pre_meta = build_preprocessor([], meta_cols)
X_train_meta_p = pre_meta.fit_transform(X_train_meta)
X_val_meta_p = pre_meta.transform(X_val_meta)

# Combined
pre_all = build_preprocessor(cpg_cols, meta_cols)
X_train_all_p = pre_all.fit_transform(X_train_all)
X_val_all_p = pre_all.transform(X_val_all)
```

```python
from sklearn.linear_model import LinearRegression
from src.functions import bootstrap_evaluation

models = {
    "OLS_CpG": (X_train_cpg_p, X_val_cpg_p),
    "OLS_Metadata": (X_train_meta_p, X_val_meta_p),
    "OLS_All": (X_train_all_p, X_val_all_p)
}

results = {}

for name, (X_tr, X_v) in models.items():
    
    model = LinearRegression()
    model.fit(X_tr, y_train)

    preds = model.predict(X_v)

    res, boot = bootstrap_evaluation(y_val, preds)
    
    results[name] = res
```
